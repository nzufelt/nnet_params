```python
# Needed to import FFNN class:
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
from ffnn import FFNN
# Rest of the imports:
import numpy as np
from sklearn.datasets import make_moons
import theano
import theano.tensor as T
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create the data
I = np.identity(2)
X_data,y = make_moons(200,noise=.15)
y_data = np.array([I[value] for value in y])

# Center the data
X_data[:,0] -= X_data[:,0].mean()
X_data[:,1] -= X_data[:,1].mean()

# Plot the data
plt.scatter(X_data[:,0], X_data[:, 1], c=y[:], s=50,
           cmap="RdBu", vmin=-.2, vmax=1.2,
           edgecolor="white", linewidth=1)
plt.show()
```

```python
nn = FFNN(2,[250,45],2,reg=.0001,epochs=10000,noise_scale=15.0)

fit_score = nn.fit(X_data,y_data)
for fit_line in fit_score:
    print(fit_line)
print(nn.accuracy(X_data,y_data))
```

```python
#plt.scatter(X_data[:,0], X_data[:, 1],c=y*255,cmap='RdBu')
#plt.show()

xx, yy = np.mgrid[-2.5:2.5:.01, -1.5:1.5:.01]
grid = np.c_[xx.ravel(), yy.ravel()]
probs = nn.predict(grid).reshape(xx.shape)

f, ax = plt.subplots(figsize=(10, 7))
contour = ax.contourf(xx,yy, probs, 25, cmap="RdBu",
                      vmin=0, vmax=1)

ax.scatter(X_data[:,0], X_data[:, 1], c=y[:], s=50,
           cmap="RdBu", vmin=-.2, vmax=1.2,
           edgecolor="white", linewidth=1)
plt.show()
```
